```python
import numpy as np
import matplotlib.pyplot as plt
```

# Finite Differences

```python
x = np.linspace(-1,1,101)

def f(x):
    return x
def g(x):
    return x**2
def h(x):
    return x**3
def k(x):
    return x**4
```

```python
plt.clf()

plt.subplot(2,2,1)
plt.plot(x, f(x))

plt.subplot(2,2,2)
plt.plot(x, g(x))

plt.subplot(2,2,3)
plt.plot(x, h(x))

plt.subplot(2,2,4)
plt.plot(x, k(x))

plt.show()
```
